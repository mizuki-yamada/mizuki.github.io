```python
import numpy as np
import math
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#初期条件
T_init=0.0
T_max=1.0

L=40.0
U=1.0

delta_x=0.1
delta_t=0.0001
```

```python
#xの範囲
x=np.arange(0.0,L,delta_x)
#xの繰り返し回数
N_step=int(L/delta_x)

#温度分布T
T_before1=[0.0]*N_step

#係数計算
Pe = 0.1
D1=U*L/Pe
c1=delta_t/(Pe*delta_x**2)
c2=(U*delta_t)/(2*delta_x)

#Tの変化
T_updated1=T_before1.copy()
T_updated1[0]=1.0
T_updated1[-1]=0.0

#経過時間
t1=0.0

inf=float('inf')

while(np.max(np.abs(np.array(T_updated1)-np.array(T_before1)>=0.0001))):
  t1+=delta_t
  T_before1=T_updated1.copy()
  T_updated1[0]=1.0
  T_updated1[-1]=0.0

  for i in range(1,N_step-1):
    #時刻tにおける各地点の温度を計算
    T_updated1[i]=(1-2*c1)*T_before1[i]+(c1+c2)*T_before1[i-1]+(c1-c2)*T_before1[i+1]
    if T_updated1[i]==inf:
      break    
print(T_updated1)
print(t1)
```
